```python
import os
import json
import pandas

data_file_name = 'statistics.json'

# загружаем все json данные
with open(data_file_name, 'r') as fp:
    data_list = list(json.load(fp))

len(data_list)
```

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# переводим данные из файла в dataFrame
base_df = pd.DataFrame(data_list)
base_df['original_video_size'] = base_df['original_video_size'].astype(str)
base_df['patch_size'] = base_df['patch_size'].apply(lambda x: x[0])
base_df.head()
```

```python
base_df_avrg = base_df[['original_video_size', 'model_name', 'patch_size','productivity_ratio']].groupby(by=['original_video_size', 'model_name', 'patch_size']).mean().reset_index()
base_df_avrg['productivity_ratio'] = base_df_avrg['productivity_ratio'].round(2)
base_df_avrg = base_df_avrg.pivot(index=['original_video_size', 'model_name'], columns='patch_size', values='productivity_ratio')

base_df_avrg.head(50)
```

```python
base_df_fps = base_df[['original_video_size', 'model_name', 'patch_size','max_possible_FPS']].groupby(by=['original_video_size', 'model_name', 'patch_size']).mean().reset_index()
base_df_fps['max_possible_FPS'] = base_df_fps['max_possible_FPS'].round(0).astype(int)
base_df_fps = base_df_fps.pivot(index=['original_video_size', 'model_name'], columns='patch_size', values='max_possible_FPS')

base_df_fps.head(50)
```
